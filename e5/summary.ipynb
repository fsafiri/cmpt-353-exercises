```python
import pandas as pd

def csv_reader(paths):
    for i in paths:
        data = pd.read_csv(i)
        std = data.std().to_string()
        mean = data.mean().to_string()
        min_val = data.min().to_string()
        max_val = data.max().to_string()
        corr=data.corr()
        corr=corr.loc['x', 'y']
        
        print(f'Description of file: {i}\n')
        print(f'Mean:\n{mean}\n')
        print(f'Standard Deviation:\n{std}\n')
        print(f'Minimum:\n{min_val}\n')
        print(f'Maximum:\n{max_val}\n')
        print(f'Correlation coefficient:\n{corr.round(2)}\n')
    
        print('-------------------------------')


paths=['data-1.csv','data-2.csv','data-3.csv','data-4.csv','data-5.csv','data-6.csv']
csv_reader(paths)
#0.9503256954375487
            
```
